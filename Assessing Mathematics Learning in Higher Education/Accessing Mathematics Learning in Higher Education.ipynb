```python
# Load the dataset
file_path = 'MathE dataset (4).csv'
data = pd.read_csv(file_path, encoding='ISO-8859-1')
```

```python
# Check if 'Question Level' column exists and display unique levels and their counts
if 'Question Level' in data.columns:
    question_levels = data['Question Level'].value_counts()
    print("Question Levels and their Counts:")
    print(question_levels)
else:
    print("Column 'Question Level' not found in the dataset.")
```

```python
# Check if 'Question Level' column exists and display unique levels and their counts
if 'Question Level' in data.columns:
    student_country = data['Student Country'].value_counts()
    print("Student Country and their Counts:")
    print(student_country)
else:
    print("Column 'Student Country' not found in the dataset.")
```

```python
# Check the distribution of Question Levels
if 'Question Level' in data.columns:
    distribution = data['Question Level'].value_counts()
    print("Distribution of Question Levels:")
    print(distribution)
    max_count = distribution.max()
    min_count = distribution.min()
    
    if max_count / min_count > 1.5:
        print("The dataset is imbalanced.")
    else:
        print("The dataset is balanced.")
else:
    print("Column 'Question Level' not found in the dataset.")
```

```python
# Check and plot the distribution of 'Question Level'
if 'Question Level' in data.columns:
    plt.figure(figsize=(8, 8))
    sns.countplot(data=data, x='Question Level', order=['Basic', 'Intermediate', 'Advanced'])
    plt.title('Distribution of Question Levels')
    plt.xlabel('Question Level')
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.show()
else:
    print("Column 'Question Level' not found in the dataset.")
```

```python
# Check and plot the distribution of 'Student Country'
if 'Student Country' in data.columns:
    plt.figure(figsize=(8, 8))
    sns.countplot(data=data, x='Student Country')
    plt.title('Distribution of Student Country')
    plt.xlabel('Student Country')
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.show()
else:
    print("Column 'Student Country' not found in the dataset.")
```

```python
# Pie Chart - Proportion of Topics
plt.figure(figsize=(8, 8))
data['Topic'].value_counts().nlargest(8).plot.pie(autopct='%1.1f%%', colors=sns.color_palette("pastel"))
plt.title('8 Topics Proportion')
plt.ylabel('')
plt.show()
```

```python
# Histogram - Distribution of numerical data
numerical_columns = data.select_dtypes(include=['number']).columns
if len(numerical_columns) > 0:
    for col in numerical_columns:
        plt.figure(figsize=(8, 6))
        sns.histplot(data[col], kde=True, bins=40, color='blue')
        plt.title(f'Distribution of {col}')
        plt.xlabel(col)
        plt.ylabel('Frequency')
        plt.show()
```

```python
# Scatter Plot - Relationships between numerical variables
if len(numerical_columns) > 1 :
    plt.figure(figsize=(8, 6))
    sns.scatterplot(data=data, x=numerical_columns[0], y=numerical_columns[1], hue='Question Level', palette='deep')
    plt.title(f'{numerical_columns[0]} vs {numerical_columns[1]}')
    plt.xlabel(numerical_columns[0])
    plt.ylabel(numerical_columns[1])
    plt.show()
```

```python
# Identify numerical and categorical columns
numerical_columns = data.select_dtypes(include=['number']).columns
categorical_columns = data.select_dtypes(include=['object', 'category']).columns
```

```python
# Check if the relevant columns exist
if 'Student Country' in data.columns and 'Topic' in data.columns:
    plt.figure(figsize=(12, 8))
    sns.countplot(data=data, y='Topic', hue='Student Country', palette='Set2')
    plt.title('Topics Asked by Students from Different Countries', fontsize=16)
    plt.xlabel('Count', fontsize=14)
    plt.ylabel('Topic', fontsize=14)
    plt.legend(title='Student Country', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()
else:
    print("Columns 'Student Country' and/or 'Topic' are missing in the dataset.")
```

```python
if len(numerical_columns) > 1 and len(categorical_columns) > 0:
    plt.figure(figsize=(8, 6))
    sns.scatterplot(data=data, x=numerical_columns[0], y=numerical_columns[1], hue=categorical_columns[0], palette='Set1')
    plt.title(f'Scatter Plot: {numerical_columns[0]} vs {numerical_columns[1]} by {categorical_columns[0]}')
    plt.xlabel(numerical_columns[0])
    plt.ylabel(numerical_columns[1])
    plt.legend(title=categorical_columns[0])
    plt.show()
```

```python
# Check if 'Student Country' column exists
if 'Student Country' in data.columns:
    country_counts = data['Student Country'].value_counts()
    selected_palette = 'Set2'
    plt.figure(figsize=(12, 6))
    sns.barplot(x=country_counts.index, y=country_counts.values, palette=selected_palette)
    plt.title('Number of Students by Country')
    plt.xlabel('Country')
    plt.ylabel('Number of Students')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
else:
    print("Column 'Student Country' not found in the dataset.")
```

```python
# Check if 'Student Country' column exists
if 'Student Country' in data.columns:
    country_counts = data['Student Country'].value_counts()
    top_countries = country_counts.head(5)
    other_countries = pd.Series({'Others': country_counts.iloc[5:].sum()})
    pie_data = pd.concat([top_countries, other_countries])
    plt.figure(figsize=(8, 8))
    pie_data.plot.pie(
        autopct='%1.1f%%', 
        startangle=90, 
        colors=plt.cm.Paired.colors,
        explode=[0.1 if i == 0 else 0 for i in range(len(pie_data))]
    )
    plt.title('Top Countries by Number of Students')
    plt.ylabel('')
    plt.tight_layout()
    plt.show()
else:
    print("Column 'Student Country' not found in the dataset.")
```

```python
# Plotting the distribution of questions by subtopic
plt.figure(figsize=(10, 6))
sns.countplot(x='Subtopic', data=data, palette='Set1')
plt.title('Number of Questions by Subtopic')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Plotting the most common keywords
if 'Keywords' in data.columns:
    keywords = data['Keywords'].dropna().str.split(',').explode().str.strip()
    keyword_counts = keywords.value_counts().head(10)
    plt.figure(figsize=(10, 6))
    sns.barplot(x=keyword_counts.index, y=keyword_counts.values, palette='coolwarm')
    plt.title('Top 10 Most Common Keywords')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```
